#### 2d) Display a pie chart of the total sales of all Grand Theft Auto games combined in North America, Europe, Japan, and other sales.

```python
# calculating the percentage sales
gta_games = df_refresh[df_refresh['Name'].str.contains("Grand Theft Auto", case=False, na=False)]
sum_of_sales = gta_games['Global Sales'].sum()

treshold = 5

gta_games.loc[:, 'Percentage Sales'] = round((gta_games["Global Sales"] / sum_of_sales) * 100, 2)

gta_games_sorted = gta_games.sort_values(by='Percentage Sales', ascending= False)

gta_games_sorted.head()
#group values less than 5% into Category of Games called MISC
threshold = 5
Sorted_names = []
categories = []
for index,rows in gta_games_sorted.iterrows():

    if (rows['Percentage Sales']) < threshold:
        categories.append('Misc:>5%')
    else:
        categories.append(rows['Name'])
gta_games_sorted['Names_for_pie'] = categories
# gta_games_sorted.head()
```

```python
gta_sales_grouped = gta_games_sorted.groupby('Names_for_pie')['Percentage Sales'].sum().reset_index()
# gta_sales_grouped.tail(6)
```

```python
# Plotting the pie chart
plt.figure(figsize=(12, 10))
plt.pie(gta_sales_grouped['Percentage Sales'], labels=gta_sales_grouped['Names_for_pie'], autopct='%1.2f%%', startangle=60,labeldistance=1.05)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
#plt.savefig("pie_chart.png")
plt.show()
```
